```python
import spacy

sentence = "Elon Musk och Steven Hawking pratar om AI på Migrationsverket"
nlp = spacy.load("sv_core_news_lg")
doc = nlp(sentence)
print(doc.text)
for token in doc:
    print(token.text, token.pos_, token.dep_)

print("entities:")
for ent in doc.ents:
    print(ent.text, ent.label_)
```
